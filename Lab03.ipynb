```python
def primary(list1):    
    variable1=0
    variable2=0
    for i in list1: #MEAN
        variable1+=1
        variable2+=i
        mean=variable2/variable1
    print("Mean: ",mean)


def primary2(list1):
    variable3=list1[0]
    for j in list1:
        if j>variable3:
            variable3=j
    print("Maximum: ",variable3)

def primary3(list1):
    variable4=list1[0]
    for j in list1:
        if j<variable4:
            variable4=j
    print("Minimum: ",variable4)

def primary4(list1):
    variable4=list1[0]
    variable3=list1[0]
    for j in list1:
        if j<variable4:
            variable4=j
        for l in list1:
            if l>variable3:
                variable3=l  # return variable3
                               # return variable4
    
    print("Range: ",variable3-variable4)




```
